```python
from sklearn.model_selection import train_test_split
```

## Split the dataset into train and test

```python
x=data.drop(['Outcome'],axis=1)
y=data['Outcome']
```

```python
print(x)
```

```python
print(y)
```

```python
xTrain,xTest,yTrain,yTest=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
xTest.shape
```

```python
xTrain.shape
```

```python
print(xTrain.dtypes)
```

```python
print(yTrain.dtypes)
```

## Necessary Packages for Build the Model 

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report, accuracy_score, precision_score, f1_score, recall_score
```

```python
model=GaussianNB()
```

## Fit the model

```python
model.fit(xTrain,yTrain)
```

## Predict on the test data

```python
predict=model.predict(xTest)
```

```python
print("Predicted Value:",predict)
```

## Confusion Matrix

```python
from sklearn import metrics
```

```python
cm = metrics.confusion_matrix(yTest, predict)
print(cm)
```

```python
import seaborn as sns
sns.heatmap(cm, annot=True, fmt='d', cmap='YlGnBu' )
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('Actual Label')
plt.show()
```

## Classificiation Report

```python
print(classification_report(yTest, predict))
```

## Accuracy , Precision , Recall , F1-Score

```python
# Calculate accuracy
accuracy = accuracy_score(yTest, predict)
print(f'Accuracy: {accuracy:.4f}')

# Calculate precision
precision = precision_score(yTest, predict)
print(f'Precision: {precision:.4f}')

# Calculate recall
recall = recall_score(yTest, predict)
print(f'Recall: {recall:.4f}')

# Calculate F1 score
f1 = f1_score(yTest, predict)
print(f'F1 Score: {f1:.4f}')
```
